## Decision Tree data weather dengan menggunakan python

```python
from sklearn import tree 
import pandas as pd 
import pydotplus # untuk membuat tree
from IPython.display import Image  # untuk menampilkan tree
```

membuat tabel data weather seperti diatas:

```python
tennis_df = pd.DataFrame()

#menambah atribut outlook
tennis_df['Outlook'] = ['sunny', 'sunny', 'overcast', 'rainy', 'rainy', 'rainy', 
                     'overcast', 'sunny', 'sunny', 'rainy', 'sunny', 'overcast',
                     'overcast', 'rainy']

#menambah atribut temp.
tennis_df['Temperature'] = ['hot', 'hot', 'hot', 'mild', 'cool', 'cool', 'cool',
                         'mild', 'cool', 'mild', 'mild', 'mild', 'hot', 'mild']

#menambah atribut humidity
tennis_df['Humidity'] = ['high', 'high', 'high', 'high', 'normal', 'normal', 'normal',
                      'high', 'normal', 'normal', 'normal', 'high', 'normal', 'high']

#menambah atribut windy
tennis_df['Windy'] = ['false', 'true', 'false', 'false', 'false', 'true', 'true',
                   'false', 'false', 'false', 'true', 'true', 'false', 'true']

#play tennis
tennis_df['Play'] = ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 
                  'yes', 'yes', 'no']


#Print/show the new data
print(tennis_df)
```

mengubah data menjadi binary

```python
one_hot_data = pd.get_dummies(tennis_df[ ['Outlook', 'Temperature', 'Humidity', 'Windy'] ])
one_hot_data
```

Klasifikasi pohon keputusan:

```python
# The decision tree classifier.
clf = tree.DecisionTreeClassifier()
clf_train = clf.fit(one_hot_data, tennis_df['Play'])
```

Menampilkan decision tree:

```python
print(tree.export_graphviz(clf_train, None))

dot_data = tree.export_graphviz(clf_train, out_file=None, feature_names=list(one_hot_data.columns.values), 
                                class_names=['Not_Play', 'Play'], rounded=True, filled=True) 

graph = pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```
